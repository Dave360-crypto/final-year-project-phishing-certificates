```python
port = "443"
url = "google.com"
```

```python
cert = OpenSSL.crypto.load_certificate(
      OpenSSL.crypto.FILETYPE_PEM,
      ssl.get_server_certificate(("google.com","443"))
)
```

```python
pemcert = OpenSSL.crypto.dump_certificate(OpenSSL.crypto.FILETYPE_PEM, cert)
```

```python
f = open('cert3.pem','wb')
f.write(pemcert)
f.close()
```
